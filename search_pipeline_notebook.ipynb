# Поиск по файлам в директории
Этот блокнот выполняет поиск по всем файлам в указанной директории и выводит совпадения с заданным запросом.

```python
import os

def search_in_directory(query, directory_path):
    results = []
    
    # Проход по всем файлам в директории
    for filename in os.listdir(directory_path):
        file_path = os.path.join(directory_path, filename)
        
        # Проверка, что это файл, а не папка
        if os.path.isfile(file_path):
            try:
                # Открытие файла и чтение содержимого
                with open(file_path, 'r', encoding='utf-8') as file:
                    content = file.read()
                    
                    # Поиск запроса в содержимом файла
                    if query in content:
                        # Сохранение названия файла и содержимого
                        results.append({
                            'file': filename,
                            'content': content
                        })
            except Exception as e:
                print(f"Не удалось прочитать файл {filename}: {e}")
    
    # Возвращение результатов
    return results

# Пример использования:
# directory_path = '/path/to/your/directory'
# query = 'your search term'
# search_results = search_in_directory(query, directory_path)
# print(search_results)
```
